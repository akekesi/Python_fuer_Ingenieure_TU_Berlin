# Aufgabe:  
**Schreiben Sie eine Funktion, die alle Pythagoreischen Tripel - also drei nat체rliche Zahlen a, b, c, sodass das Dreieck mit den Seitenl채ngen a, b und c rechtwinklig ist - berechnet und als Liste von Tupeln zur체ckgibt , wobei die Hypotenuse kleiner als ein 체bergebener Wert ist.**

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def pythagoreischen_tripel(c):
    """
    Return alle Pythagoreischen Tripel fuer Hypotenuse c
    Args:
        c: Hypotenuse, ganze Zahl > 0
    Retruns:
        Liste von Tupeln (L[(a1,b1,c),(a2,b2,c),...]) von Pythagoreischen Tripeln
        a > 0 ganze Zahl
        b > 0 ganze Zahl
        c > 0 ganze Zahl
        a^2 + b^2 = c^2
    """
    L = []
    for a in range(1,c):
        b = (c**2 - a**2)**0.5
        if b-int(b) < 10**(-12):
            L.append((a,int(b),c))
    return L
```

```python
# Test
c = 125
L = pythagoreischen_tripel(c)
print(f'Pythagoreischen Tripel fuer c={L[0][-1]:2.0f}:\n{L}')
```

```python
def pythagoreischen_tripel_plot(L):
    """
    Plot Pythagoreischen Tripel
    """
    plt.figure(figsize=(9,9))
    
    c = L[-1][-1]
    
    # Kreis mit Radius = c
    x = np.arange(0,c,0.1)
    plt.plot(125*np.sin(x),125*np.cos(x),'k--')
    
    # Dreiecke
    for l in L:
        a,b,_ = l
        P = [[0,a,a,0],[0,0,b,0]]
        plt.plot(P[0],P[1],label=l,linewidth=3)
        plt.legend(loc='upper right', fontsize=15)

    plt.xlabel('a', fontsize=20)
    plt.ylabel('b', fontsize=20, rotation=0, labelpad=15)
    plt.ylim(0,c*1.1)
    plt.xlim(0,c*1.1)
    plt.title(f'alle Pythagoreischen Tripel fuer Hypotenuse {c}', fontsize=20)
    plt.grid(True) # grid lines
```

```python
L = pythagoreischen_tripel(125)
pythagoreischen_tripel_plot(L)
```

```python
def pythagoreischen_tripel_alle(C):
    """
    Return alle Pythagoreischen Tripel fuer Hypotenuse c <= C
    Args:
        C: max. Hypotenuse, ganze Zahl > 0
    Retruns:
        Liste von Listen von Tupeln von Pythagoreischen Tripeln
        a > 0 ganze Zahl
        b > 0 ganze Zahl
        c > 0 ganze Zahl
        a^2 + b^2 = c^2
    """
    L = []
    for c in range(C+1):
        L += pythagoreischen_tripel(c)
    return L
```

```python
C = 30
pythagoreischen_tripel_alle(C)
```
