**3) Funktion auf Zahlenreihe**

  * Schreiben Sie eine Funktion, die einen Vektor übergeben bekommt und dessen quadrierte Einträge alternierend addiert bzw. subtrahiert. *(3 Punkte)*
  * Die Rechnung lautet für einen Vektor $\pmb{b}$:
$$
{b_1}^2 - {b_2}^2 + {b_3}^2 - {b_4}^2 + {b_5}^2 \ldots
$$

  * Die Funktion soll das skalare Ergebnis der Rechnung zurückgeben. *(1 Punkt)*
  * Verwenden Sie NumPy-Arrays und verzichten Sie nach Möglichkeit auf if-Klauseln und Schleifen. *(1 Punkt)*
  * Wenden Sie die Funktion auf den folgenden Vektor an und geben Sie das Ergebnis der Rechnung aus: *(1 Punkt)*

$~~~~~\pmb{b}=~\big($ `1.2, 2.8, -3.8, -1.8, -1.1, 0.1, -2.9, 3.3, -2.4, 2.7, -2.0, -1.5, 1.3, -2.8` $\big)$


```python
# Hier eigenen Code schreiben ...
def alt_quad_schleife(V):
    """
    mit Schleife
    V   = [b1, b2, b3, b4, b5, ...]
    res = b1^2 - b2^2 + b3^2 - b4^2 + b5^2 ...
    """
    res = 0
    vorz = 1
    for v in V:
        res += vorz*v**2
        vorz *= -1
    return res    
```

```python
# Hier eigenen Code schreiben ...
def alt_quad(V):
    """
    ohne Schleife
    V   = [b1, b2, b3, b4, b5, ...]
    res = b1^2 - b2^2 + b3^2 - b4^2 + b5^2 ...
    """
    a = V**2
    return np.sum(a[::2])-np.sum(a[1::2])
```

```python
# Test
#V = np.array([1,2,3,4])
#V = np.array([1,1,2,3])
V = np.array([1.2, 2.8, -3.8, -1.8, -1.1, 0.1, -2.9, 3.3, -2.4, 2.7, -2.0, -1.5, 1.3, -2.8])
print(f"mit Schleife:\n{float(alt_quad_schleife(V)):.3}")
print("\n")
print(f"ohne Schleife:\n{float(alt_quad(V)):.3}")
```

```python
# Hier eigenen Code schreiben ...
# [min, max)
x_min = -10
x_max = 25
#x = np.sort(np.random.uniform(x_min,x_max,size=126))
x = np.linspace(x_min,x_max-1,126)
y = 17*np.sin(1.3*x)
Y = np.reshape(y,(7,-1))
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.plot(x,y)
plt.axis('equal')
plt.xticks(np.arange(-10,25,10))
plt.yticks(np.arange(-15,20,5))
plt.xlabel("x / m")
plt.ylabel("y / m")
plt.grid()

plt.subplot(1,3,2)
plt.imshow(Y)
plt.xticks(np.arange(0,20,5))
plt.yticks(np.arange(0,10,5))
plt.colorbar()

plt.subplot(1,3,3)
n = 3
_,N = Y.shape
s = np.arange(44,44-N*2,-2)
for i,yy in enumerate(Y):
    if i == n:
        continue
    plt.scatter(Y[n,:],yy,s=s,label=f"{i+1}. Zeile")
plt.legend(loc='lower right')
plt.xticks(np.arange(-10,20,10))
plt.yticks(np.arange(-15,20,5))

plt.show()
```
